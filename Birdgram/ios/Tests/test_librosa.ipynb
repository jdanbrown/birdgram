# Test ios librosa vs. py

```python
from notebooks import *
figsize(aspect=1/12)
def sigp(x):
    pp((x.shape, x[:7]))
def sig(x):
    sigp(x)
    plt.plot(x); plt.show()
def mat(X, rows=10, cols=10, **show_img_kwargs):
    pp(X.shape)
    pp(X[:rows, :cols])
    show_img(X, **show_img_kwargs)
```

```python
# filters.mel
sr     = 22050  # (sample_rate)
n_fft  = 128
n_mels = 40
M = librosa.filters.mel(sr=sr, n_fft=n_fft, n_mels=n_mels)
print(json.dumps(dict(sr=sr, n_fft=n_fft, n_mels=n_mels, M=M.tolist())))  # To copy to .swift notebook
mat(M, rows=32, cols=12, scale=8)
# pp(M.shape)
# pp(M[:21, :12])
# show_img(M, scale=8)
```

```python
# fft_frequencies
sr    = 22050  # (sample_rate)
n_fft = 32
f = librosa.fft_frequencies(sr=sr, n_fft=n_fft)
print(json.dumps(dict(sr=sr, n_fft=n_fft, f=f.tolist())))  # To copy to .swift notebook
sig(f)
```

```python
# mel_frequencies
n_mels = 40
fmin   = 0
fmax   = 22050
f = librosa.mel_frequencies(n_mels=n_mels, fmin=fmin, fmax=fmax)
print(json.dumps(dict(n_mels=n_mels, fmin=fmin, fmax=fmax, f=f.tolist())))  # To copy to .swift notebook
sig(f)
```

```python
# hz_to_mel
x = np.random.rand(10)
y = librosa.hz_to_mel(x)
print(json.dumps(dict(x=x.tolist(), y=y.tolist())))  # To copy to .swift notebook
sig(x)
sig(y)
```

```python
# mel_to_hz
x = np.random.rand(10)
y = librosa.mel_to_hz(x)
print(json.dumps(dict(x=x.tolist(), y=y.tolist())))  # To copy to .swift notebook
sig(x)
sig(y)
```

```python
# power_to_db
X = np.random.rand(3,4)
Y = librosa.power_to_db(X)
print(json.dumps(dict(X=X.tolist(), Y=Y.tolist())))  # To copy to .swift notebook
mat(X, scale=10)
mat(Y, scale=10)
```
