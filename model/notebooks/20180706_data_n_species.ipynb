```python
def show_n_recs_per_species(geo_filter):
    n_recs_per_species = (xc.metadata
        [lambda df: pd.notnull(df.species)]
        [geo_filter]
        .pipe(df_remove_unused_categories).com_name.sort_values()
        .value_counts().reset_index().rename(columns={'index': 'com_name', 'com_name': 'n_recs'})
    )
    display(
        (n_recs_per_species
            .reset_index().rename(columns={'index': 'n_species', 'n_recs': '≥ n_recs'})
            .pipe(ggplot, aes(x='n_species', y='≥ n_recs'))
            + geom_line()
            + scale_x_continuous(breaks=lambda lims: np.arange(0, (lims[1] // 10 + 1) * 10, 10))
            + scale_y_continuous(breaks=lambda lims: np.arange(0, (lims[1] // 10 + 1) * 10, 10))
            + theme_figsize('inline_short')
        ),
        [
            '%s species where n_recs ≥ %s' % (len(n_recs_per_species[lambda df: df.n_recs >= n_recs]), n_recs)
            for n_recs in [20, 15, 10, 5, 1]
        ],
        n_recs_per_species,
    )
```

```python
show_n_recs_per_species(
    lambda df: (df.country == 'United States') & df.locality.str.endswith('California'),
)
```
